```python
import graphlab
```

# Load the CIFAR-10 dataset

```python
image_train = graphlab.SFrame('image_train_data/')
```

```python
image_train.head()
```

# Train a nearestneighbors model for retrieving images using deep features

```python
knn_model = graphlab.nearest_neighbors.create(image_train, label='id', features=['deep_features'])
```

# Use image retrieval model with deep features to find similar images

```python
graphlab.canvas.set_target('ipynb')
cat = image_train[18:19]
cat['image'].show()
```

```python
knn_model.query(cat)
```

```python
def get_images_ids(query_result):
    return image_train.filter_by(query_result['reference_label'], 'id')

```

```python
cat_neighbors = get_images_ids(knn_model.query(cat))
```

```python
cat_neighbors.show()
```

```python
car = image_train[8:9]
```

```python
car['image'].show()
```

```python
car_neighbors = get_images_ids(knn_model.query(car))
```

```python
car_neighbors['image'].show()
```

## Task 4

```python
image_test_dog = image_train[image_train['label'] == 'dog']
image_test_auto = image_train[image_train['label'] == 'automobile']
image_test_cat = image_train[image_train['label'] == 'cat']
image_test_bird = image_train[image_train['label'] == 'bird']
```

```python
dog_cat_neighbors = cat_model.query(image_test_dog, k=1)
dog_bird_neighbors = bird_model.query(image_test_dog, k=1)
dog_auto_neighbors = auto_model.query(image_test_dog, k=1)
dog_dog_neighbors = dog_model.query(image_test_dog, k=1)
```

```python
dog_dog_neighbors
```

```python
dog_auto_neighbors
```

```python
dog_distances = graphlab.SFrame()
dog_distances['dog-dog'] = dog_dog_neighbors['distance']
dog_distances['dog-cat'] = dog_cat_neighbors['distance']
dog_distances['dog-auto'] = dog_auto_neighbors['distance']
dog_distances['dog-bird'] = dog_bird_neighbors['distance']
```

```python
dog_distances.head()
```

```python
def is_dog_correct(row):
    if row['dog-dog'] < row['dog-cat']  and  row['dog-dog'] < row['dog-bird']  and  row['dog-dog'] < row['dog-auto']:
        return 1
    return 0
```

```python
dog_distances['correct'] = dog_distances.apply(is_dog_correct)
```

```python
print sum(dog_distances['correct'])
print len(dog_distances)
```
